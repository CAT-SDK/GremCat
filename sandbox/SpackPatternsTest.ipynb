### The details on generating this heatmap
What follows are the steps encapsulated in the patterns.view_top_N_map(N) function.

```python
heat_obj = patterns.make_file_developer_df()
```

```python
heat_obj.index.name = 'File'
```

```python
# import matplotlib.pyplot as plt
# import seaborn as sb
# fig, ax = plt.subplots(figsize=(12,8))
# sb.heatmap(heat_obj, annot=True, linewidths=.5, cmap='icefire')
# if patterns.year is None:
#     plt.title('Overall developers vs files (' + str(self.project_name) + ')')
# else:
#     plt.title(str(self.year) + ' : developers vs files (' + self.project_name + ')')
```

```python
sorted_developers = heat_obj.sum(axis='rows').sort_values(ascending=False)
top_developers = sorted_developers.head(20)
hot_developers = heat_obj[top_developers.index]
hot_developers.head(20)
```

```python
top_files = heat_obj.sum(axis='columns').sort_values(ascending=False).head(20)
top_files.head(20)
```

```python
hot_files = heat_obj.iloc[heat_obj.index.isin(top_files.to_dict().keys())]
# drop 0 columns
hot_files = hot_files.loc[:, (hot_files != 0).any(axis=0)]
hot_files.head(20)
```

```python
sorted_full_dev_list = list(sorted_developers.to_dict().keys())
sorted_dev_list = []
for dev in sorted_full_dev_list: 
    if dev in hot_files.columns:
        sorted_dev_list.append(dev)

sorted_hot_files = hot_files[sorted_dev_list[:20]]  # rearrange columns sorted by total dev. contributions, show only first 20 dev. columns
sorted_hot_files
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
fig, ax = plt.subplots(figsize=(20,20))         # Sample figsize in inches
sns.set(font_scale=1.5) 
sns.heatmap(sorted_hot_files, annot=True, linewidths=.5, ax=ax, fmt='g')
```
