total changed lines over time

```python
df = vis.plot_project_locc_line()
```

```python
import seaborn as sns
sns.set(font_scale=1.5)
with sns.axes_style("whitegrid"):
    g = sns.relplot(data=df, x="datetime", y="log_locc_diff", size="change-size", hue="change-size", sizes=(50, 500),
                    height=6, aspect=1.5, kind="scatter")
    g.ax.set_xlabel('Date')
    g.ax.set_ylabel('log10(lines added - lines removed)')
    g.fig.autofmt_xdate()
    g.fig.show()
df.head()
```
